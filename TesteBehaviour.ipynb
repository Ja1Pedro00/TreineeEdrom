```python
import matplotlib.pyplot as plt

origem = [0, 0]
x = int (input('Digite o valor de X (entre -9 e 9)= '))
y = int (input('Digite o valor de Y (entre -4 e 4 = '))
posicao = [x, y]
ax = plt.axes()
ax.set_xlim(-10, 10)
ax.set_ylim(-5, 5)

ax.grid(which = 'both', color = 'grey', linewidth = 1, linestyle = '-', alpha = 0.2)
plt.plot([-10, 10], [0, 0], color = 'blue', linewidth = 1)
plt.plot([0, 0], [-5, 5], color = 'blue', linewidth = 1)
plt.xlabel('largura')
plt.ylabel('altura')
plt.plot(x, y, marker = 'o', color = 'green')
plt.plot([x, 0], [0, 0], color = 'red', linewidth = 1)
plt.plot([0, 0], [0, y], color = 'red', linewidth = 1)

ax.annotate(f"{x}", (x, -0.45), fontsize = 10)
ax.annotate(f"{y}", (-0.45, y), fontsize = 10)


plt.show()
```
